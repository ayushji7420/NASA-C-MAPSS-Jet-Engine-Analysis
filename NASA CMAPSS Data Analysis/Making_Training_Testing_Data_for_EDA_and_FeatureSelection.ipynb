# Processing the Train Data

```python
import pandas as pd
import numpy as np
```

```python
Columns = ["ID", "Cycle", "OpSet1", "OpSet2", "OpSet3", "SM1", "SM2", "SM3", "SM4", "SM5", "SM6", "SM7", "SM8", "SM9", "SM10", "SM11", "SM12", "SM13", "SM14", "SM15", "SM16", "SM17", "SM18", "SM19", "SM20", "SM21"]

#For FD001
raw_data1 = pd.read_csv("Dataset/train_FD001.txt", sep =' ', header=None)
raw_data1 = raw_data1.drop([26,27], axis="columns")
raw_data1.columns = Columns

#For FD002
raw_data2 = pd.read_csv("Dataset/train_FD002.txt", sep =' ', header=None)
raw_data2 = raw_data2.drop([26,27], axis="columns")
raw_data2.columns = Columns

#For FD003
raw_data3 = pd.read_csv("Dataset/train_FD003.txt", sep =' ', header=None)
raw_data3 = raw_data3.drop([26,27], axis="columns")
raw_data3.columns = Columns

#For FD004
raw_data4 = pd.read_csv("Dataset/train_FD004.txt", sep =' ', header=None)
raw_data4 = raw_data4.drop([26,27], axis="columns")
raw_data4.columns = Columns

raw_data1
```

```python
RUL = []
for m in raw_data1["ID"]:
    RUL.append((raw_data1[raw_data1["ID"] == m]["Cycle"]).values[-1])
raw_data1["RUL"] = RUL

RUL = []
for m in raw_data2["ID"]:
    RUL.append((raw_data2[raw_data2["ID"] == m]["Cycle"]).values[-1])
raw_data2["RUL"] = RUL

RUL = []
for m in raw_data3["ID"]:
    RUL.append((raw_data3[raw_data3["ID"] == m]["Cycle"]).values[-1])
raw_data3["RUL"] = RUL

RUL = []
for m in raw_data4["ID"]:
    RUL.append((raw_data4[raw_data4["ID"] == m]["Cycle"]).values[-1])
raw_data4["RUL"] = RUL

raw_data1
```

```python
raw_data1["RUL"] = raw_data1["RUL"].sub(raw_data1["Cycle"])
raw_data2["RUL"] = raw_data2["RUL"].sub(raw_data2["Cycle"])
raw_data3["RUL"] = raw_data3["RUL"].sub(raw_data3["Cycle"])
raw_data4["RUL"] = raw_data4["RUL"].sub(raw_data4["Cycle"])

raw_data1
```

```python
raw_data1.to_csv("RUL Prediction Data/RUL_train_FD001.csv", index=False)
raw_data2.to_csv("RUL Prediction Data/RUL_train_FD002.csv", index=False)
raw_data3.to_csv("RUL Prediction Data/RUL_train_FD003.csv", index=False)
raw_data4.to_csv("RUL Prediction Data/RUL_train_FD004.csv", index=False)
```

# Processing the Test Data

```python
Columns = ["ID", "Cycle", "OpSet1", "OpSet2", "OpSet3", "SM1", "SM2", "SM3", "SM4", "SM5", "SM6", "SM7", "SM8", "SM9", "SM10", "SM11", "SM12", "SM13", "SM14", "SM15", "SM16", "SM17", "SM18", "SM19", "SM20", "SM21"]

#For FD001
test_raw_data1 = pd.read_csv("Dataset/test_FD001.txt", sep =' ', header=None)
RUL_Data1 = pd.read_csv("Dataset/RUL_FD001.txt", sep = ' ', header=None)
RUL_Data1 = RUL_Data1.drop([1], axis='columns')
RUL_Data1.columns = ["RUL"]
test_raw_data1 = test_raw_data1.drop([26,27], axis="columns")
test_raw_data1.columns = Columns

#For FD002
test_raw_data2 = pd.read_csv("Dataset/test_FD002.txt", sep =' ', header=None)
RUL_Data2 = pd.read_csv("Dataset/RUL_FD002.txt", sep = ' ', header=None)
RUL_Data2 = RUL_Data2.drop([1], axis='columns')
RUL_Data2.columns = ["RUL"]
test_raw_data2 = test_raw_data2.drop([26,27], axis="columns")
test_raw_data2.columns = Columns

#For FD003
test_raw_data3 = pd.read_csv("Dataset/test_FD003.txt", sep =' ', header=None)
RUL_Data3 = pd.read_csv("Dataset/RUL_FD003.txt", sep = ' ', header=None)
RUL_Data3 = RUL_Data3.drop([1], axis='columns')
RUL_Data3.columns = ["RUL"]
test_raw_data3 = test_raw_data3.drop([26,27], axis="columns")
test_raw_data3.columns = Columns

#For FD004
test_raw_data4 = pd.read_csv("Dataset/test_FD004.txt", sep =' ', header=None)
RUL_Data4 = pd.read_csv("Dataset/RUL_FD004.txt", sep = ' ', header=None)
RUL_Data4 = RUL_Data4.drop([1], axis='columns')
RUL_Data4.columns = ["RUL"]
test_raw_data4 = test_raw_data4.drop([26,27], axis="columns")
test_raw_data4.columns = Columns
```

```python
RUL = []
for m in test_raw_data1["ID"]:
    RUL.append((test_raw_data1[test_raw_data1["ID"] == m]["Cycle"]).values[-1])
test_raw_data1["RUL"] = RUL

RUL = []
for m in test_raw_data2["ID"]:
    RUL.append((test_raw_data2[test_raw_data2["ID"] == m]["Cycle"]).values[-1])
test_raw_data2["RUL"] = RUL

RUL = []
for m in test_raw_data3["ID"]:
    RUL.append((test_raw_data3[test_raw_data3["ID"] == m]["Cycle"]).values[-1])
test_raw_data3["RUL"] = RUL

RUL = []
for m in test_raw_data4["ID"]:
    RUL.append((test_raw_data4[test_raw_data4["ID"] == m]["Cycle"]).values[-1])
test_raw_data4["RUL"] = RUL
```

```python
RUL = []
for index,row in test_raw_data1.iterrows():
    RUL.append(RUL_Data1.loc[int(row["ID"]-1), "RUL"])
test_raw_data1["RUL"] = test_raw_data1["RUL"] + RUL

RUL = []
for index,row in test_raw_data2.iterrows():
    RUL.append(RUL_Data2.loc[int(row["ID"]-1), "RUL"])
test_raw_data2["RUL"] = test_raw_data2["RUL"] + RUL

RUL = []
for index,row in test_raw_data3.iterrows():
    RUL.append(RUL_Data3.loc[int(row["ID"]-1), "RUL"])
test_raw_data3["RUL"] = test_raw_data3["RUL"] + RUL

RUL = []
for index,row in test_raw_data4.iterrows():
    RUL.append(RUL_Data4.loc[int(row["ID"]-1), "RUL"])
test_raw_data4["RUL"] = test_raw_data4["RUL"] + RUL
```

```python
test_raw_data1["RUL"] = test_raw_data1["RUL"].sub(test_raw_data1["Cycle"])
test_raw_data2["RUL"] = test_raw_data2["RUL"].sub(test_raw_data2["Cycle"])
test_raw_data3["RUL"] = test_raw_data3["RUL"].sub(test_raw_data3["Cycle"])
test_raw_data4["RUL"] = test_raw_data4["RUL"].sub(test_raw_data4["Cycle"])

test_raw_data1
```

```python
test_raw_data1.to_csv("RUL Prediction Data/RUL_test_FD001.csv", index=False)
test_raw_data2.to_csv("RUL Prediction Data/RUL_test_FD002.csv", index=False)
test_raw_data3.to_csv("RUL Prediction Data/RUL_test_FD003.csv", index=False)
test_raw_data4.to_csv("RUL Prediction Data/RUL_test_FD004.csv", index=False)
```
